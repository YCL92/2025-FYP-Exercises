```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification

from collections import Counter
np.random.seed(1907)

from imblearn.under_sampling import RandomUnderSampler
from imblearn.over_sampling import RandomOverSampler
from imblearn.over_sampling import SMOTE
```

## Base dataset

```python
X, y = make_classification(n_samples=100,weights=(0.9,0.1),class_sep=0.8,n_features=2,n_informative=2,n_redundant=0,random_state=1907)

plt.figure()
plt.scatter(X[y==0][:,0],X[y==0][:,1],c="red")
plt.scatter(X[y==1][:,0],X[y==1][:,1],c="blue")
plt.show()
print(Counter(y))
```

## After random under-sampling

```python
randomUnderSampler = RandomUnderSampler()
X_under, y_under = randomUnderSampler.fit_resample(X, y)
plt.figure()
plt.scatter(X_under[y_under==0][:,0],X_under[y_under==0][:,1],c="red")
plt.scatter(X_under[y_under==1][:,0],X_under[y_under==1][:,1],c="blue")
plt.show()
```

## After random over-sampling

```python
randomOverSampler = RandomOverSampler()
X_over, y_over = randomOverSampler.fit_resample(X, y)
plt.figure()
plt.scatter(X_over[y_over==0][:,0],X_over[y_over==0][:,1],c="red")
plt.scatter(X_over[y_over==1][:,0],X_over[y_over==1][:,1],c="blue")
plt.show()
print(Counter(y_over))
```

## After SMOTE

```python
smote = SMOTE(random_state=1907,sampling_strategy='minority',k_neighbors=5)

X_smote,y_smote = smote.fit_resample(X,y)
plt.figure()
plt.scatter(X_smote[y_smote==0][:,0],X_smote[y_smote==0][:,1],c="red")
plt.scatter(X_smote[y_smote==1][:,0],X_smote[y_smote==1][:,1],c="blue")
plt.show()
print(Counter(y_over))
```
