```python
import random
import matplotlib.pyplot as plt
import skimage.io as io
from skimage.transform import rotate
from skimage.util import random_noise
from skimage.exposure import adjust_gamma
```

### Load the image

```python
base_img = io.imread("TRAIN000000.png")
plt.figure()
plt.imshow(base_img)
plt.axis("off")
plt.show()
```

## Rotate the image

```python
img_rotate = rotate(base_img,angle=90)
plt.figure()
plt.imshow(img_rotate)
plt.axis("off")
plt.show()
```

### With a random angle

```python
range_angle = (-45,45)
for i in range(5):
    selected_angle = random.randint(range_angle[0],range_angle[1])
    img_rotate = rotate(base_img,angle=selected_angle)
    plt.figure()
    plt.imshow(img_rotate)
    plt.title(f"Angle={selected_angle}")
    plt.axis("off")
    plt.show()
```

## Flip horizontal

```python
img_flip_horizontal = base_img[:,::-1,:]
plt.figure()
plt.imshow(img_flip_horizontal)
plt.axis("off")
plt.show()
```

## Flip vertical

```python
img_flip_vertical = base_img[::-1,:,:]
plt.figure()
plt.imshow(img_flip_vertical)
plt.axis("off")
plt.show()
```

## Change pixels intensity

```python
#gamma>1 and the image will be darker, gamma<1 and the image will be brighter
img_gamma = adjust_gamma(base_img,gamma=2)
plt.figure()
plt.imshow(img_gamma)
plt.axis("off")
plt.title("gamma>1")
plt.show()
```

```python
img_gamma = adjust_gamma(base_img,gamma=0.5)
plt.figure()
plt.imshow(img_gamma)
plt.axis("off")
plt.title("gamma<1")
plt.show()
```

## Noise

```python
img_noisy = random_noise(base_img,var=0.1)
plt.figure()
plt.imshow(img_noisy)
plt.axis("off")
plt.show()
```
